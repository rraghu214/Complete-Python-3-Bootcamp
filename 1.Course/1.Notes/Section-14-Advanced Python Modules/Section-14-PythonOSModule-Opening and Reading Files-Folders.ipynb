### Deleting Files

1. os.unlink(path) --> deletes a file from folder.
2. os.rmdir(path) --> deletes an empty folder
3. shutil.rmtree(path) --> deletes a folder recursively

Deletions here cannot be reversed or undone.

Safe option is to use send2trash module.

### os.walk
- to list all the files and folders in a given directory

```python
os.walk(top='C:\\Users\\rragh\\OneDrive\\Complete-Python-3-Bootcamp')
```

```python
file_path = 'C:\\Users\\rragh\\OneDrive\\Complete-Python-3-Bootcamp'
```

```python
for folder,sub_folder,filename in os.walk(file_path):
    print(f"Currently in folder {folder}")
    print("\n")
    print(sub_folder)
    print("\n")
    print(f"File here is: {filename}")
```
