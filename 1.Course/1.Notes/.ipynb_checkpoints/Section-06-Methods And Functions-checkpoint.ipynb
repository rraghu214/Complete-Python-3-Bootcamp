```python
# Function to find if the word dog exists in a string.

def phrase_identifier(search_string, word):
    if search_string.lower() in word.lower():
        return(search_string + ' exists in the string - ' + word)
    else:
        return(search_string + ' doesnt exist in the string - ' + word)
```

```python
phrase_identifier('dog','dogmatic')
```

```python
phrase_identifier('ggd','adddaads')
```

```python
phrase_identifier('dog','Dogmatic')
```

```python
def phrase_identifier_2(search_string, word):
    return(search_string in word.lower())
```

```python
phrase_identifier_2('a','man')
```

```python
phrase_identifier_2('a','msn')
```

## PIG Latin Game
- if the word starts with a vowel then add 'ay' at last
- if the word does not start with a vowel then move the first letter to the end and then add 'ay'
- word --> ordway
- apple --> appleay

```python
def pig_latin(word):
    if word[0].lower() in 'aeiou':
        return_key = word + 'ay'
    else:
        return_key=word[1:] + word[0] + 'ay'
    return(return_key)
```

```python
print(pig_latin('apple'))
#pig_latin('apple')
print(pig_latin('word'))
```

 # * args & * * kwargs

arguments & key word arguments

To accept an arbitrary number of arguments & key word arguments, without having to define a bunch of positional parameter before hand.

```python
def product_sums(factor,*args):
    print(args*factor)
```

```python
product_sums(2,2,3,4,5)
```

```python
def value_picker(**kwargs):
    print(kwargs)
```

```python
value_picker(fruit = 'apple', veg = 'cabbage')
```

```python
d = {'fruit' :'apple', 'veg' : 'cabbage'}
value_picker(d)
```

This means that you send it as name value pair in the args and don't assign a dictionary. * * kwargs just consolidates it into a dictionary.

```python
def myfunc(*args, **kwargs):
    #print(args[0] + ' value is ' + kwargs['food'])
    print('{} value is {}.'.format(args[0],kwargs['food']))
    
myfunc(1,2,3,4,5,food = 'pasta', toy = 'hot wheels')
```
