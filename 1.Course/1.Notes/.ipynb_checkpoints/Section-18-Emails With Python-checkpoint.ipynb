# Emails with Python
- Need Admin privileges
- Enable 2 factor auth in Gmail.
- 

```python
import smtplib
```

```python
email_obj = smtplib.SMTP(host='smtp.gmail.com', port=587)
```

```python
email_obj.ehlo()
```

```python
email_obj.starttls()
```

```python
import getpass
```

```python
email = getpass.getpass('Email id: ')
```

```python
password = getpass.getpass('Passowrd: ')
```

```python
email_obj.login(email,password)
```

```python
from_address = email
to_address = email
subject = 'Subject: Test Email'
body = 'Have a good day, Raghu!'
msg = subject + '\n' + body
```

```python
email_obj.sendmail(from_address,to_address,msg)
```

```python
email_obj.quit()
```
