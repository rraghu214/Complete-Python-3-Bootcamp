# Modules & Packages

- We will see how to call another .py file from our .py file.
    - In this conext modules are just .py scripts that you call in another .py script
    - Packages are a collection of modules.
    - You can call one function from a different Python file by importing that frunction from the file.
    - To import the files from folders, ensure to make it treat them like a python package structure. This is achieved by creating empty file __ init__.py

### Points to consider.

1. Import the function by using the keyword like "from < pkg > import < func >"
2. To import a function inside a folder, use the keyword like "from < pkg.subfolder > import < func >
3. When creating a package, ensure to use __ init__.py to regard it as a part of python package.


# __ name__ & "__ main__"

- In python there is a built in variable called __ name__
- Whenever a python script is invoked directly, python internally assigns the internal variable __ name__ = "__ main__"
- Whenever a python script is invoked from a different python script then the variable there will not have __ main__. It will have packageName.scriptName 
-
